```python
import os
from itertools import product

special_tokens = ['<', '>']
start, end = special_tokens

# Create the directory if it doesn't exist
os.makedirs('data/shakespeare_char', exist_ok=True)

# Function to convert numbers to words up to 999
def number_to_words(n):
    ones = ["zero", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine",
            "ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen",
            "seventeen", "eighteen", "nineteen"]
    
    tens = ["", "", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]

    if n < 20:
        return ones[n]
    elif n < 100:
        tens_digit = n // 10
        ones_digit = n % 10
        if ones_digit == 0:
            return tens[tens_digit]
        else:
            return tens[tens_digit] + "-" + ones[ones_digit]
    else:
        hundreds_digit = n // 100
        remainder = n % 100
        hundreds_part = ones[hundreds_digit] + " hundred"
        if remainder == 0:
            return hundreds_part
        else:
            return hundreds_part + " " + number_to_words(remainder)

# File output
output_path = 'data/shakespeare_char/000_999.txt'

# Use a buffer to write in chunks
with open(output_path, 'w') as f:
    chunk_size = 10000
    buffer = []

    for i, j in product(range(999), range(999)):
        i_word = number_to_words(i)
        j_word = number_to_words(j)

        # Addition (keep results  999)
        sum_result = i + j
        if sum_result <= 999:
            buffer.append(f"{start}{i_word} plus {j_word} equals {number_to_words(sum_result)}{end}\n")

        # Subtraction (no negatives)
        if i >= j:
            diff_result = i - j
            buffer.append(f"{start}{i_word} minus {j_word} equals {number_to_words(diff_result)}{end}\n")

        # Multiplication (keep results  999)
        prod_result = i * j
        if prod_result <= 999:
            buffer.append(f"{start}{i_word} times {j_word} equals {number_to_words(prod_result)}{end}\n")

        # Division (no zero, integer results,  999)
        if j != 0 and i % j == 0:
            div_result = i // j
            if div_result <= 999:
                buffer.append(f"{start}{i_word} divided by {j_word} equals {number_to_words(div_result)}{end}\n")

        if len(buffer) >= chunk_size:
            f.writelines(buffer)
            buffer.clear()

    if buffer:
        f.writelines(buffer)

print(f" Synthetic verbal math dataset (0999) created at: {output_path}")
```

```python
import os

file_1 = 'data/shakespeare_char/input.txt'
file_2 = 'data/shakespeare_char/input_test.txt'

# Method 1: Compare file contents directly
def compare_files(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        # Read and compare line by line to avoid loading entire files into memory
        for line_num, (line1, line2) in enumerate(zip(f1, f2), 1):
            if line1 != line2:
                return False, f"Files differ at line {line_num}"
        
        # Check if one file is longer than the other
        extra_line1 = f1.read()
        extra_line2 = f2.read()
        if extra_line1 or extra_line2:
            return False, "Files have different lengths"
            
    return True, "Files are identical"

# Define human-readable size formatter
def human_readable_size(size_bytes):
    if size_bytes < 1024:
        return f"{size_bytes} bytes"
    elif size_bytes < 1024**2:
        return f"{size_bytes/1024:.2f} KB"
    elif size_bytes < 1024**3:
        return f"{size_bytes/(1024**2):.2f} MB"
    else:
        return f"{size_bytes/(1024**3):.2f} GB"

# Method 2: Compare file sizes first (quick check)
size1 = os.path.getsize(file_1)
size2 = os.path.getsize(file_2)

if size1 != size2:
    print(f"Files have different sizes: {file_1} ({human_readable_size(size1)}) vs {file_2} ({human_readable_size(size2)})")
else:
    print(f"Files have the same size: {human_readable_size(size1)}")
    identical, message = compare_files(file_1, file_2)
    print(message)
```

```python
import os
import pickle
import requests
import numpy as np

# download the tiny shakespeare dataset
include_special_tokens = False # whether to include the parts that make up special tokens in the character set disable if your dataset doesnt contain any e.g no < | s t a r t | >

print(f"length of dataset in characters: {len(input_data):,}")

# Create a unified character set
all_chars = sorted(list(set(input_data))) if include_special_tokens else sorted(list(set([c for c in input_data if c not in ''.join(special_tokens)])))
vocab_size = len(special_tokens) + len(all_chars)
print("Regular characters:", all_chars)
print(f"Total vocab size (characters + special tokens): {vocab_size:,}")

# Create mappings that include both individual characters and special tokens
stoi = {}
# First add special tokens
for i, token in enumerate(special_tokens):
    stoi[token] = i
# Then add all individual characters
for i, ch in enumerate(all_chars):
    stoi[ch] = i + len(special_tokens)

itos = {i: ch for ch, i in stoi.items()}

def encode(s):
    # Prioritize special tokens during encoding
    encoded = []
    i = 0
    while i < len(s):
        # Check if any special token starts at the current position
        is_special = False
        for token in special_tokens:
            if s[i:].startswith(token):
                encoded.append(stoi[token])
                i += len(token)
                is_special = True
                break
        if not is_special:
            # If not a special token, encode as individual character
            encoded.append(stoi[s[i]])
            i += 1
    return encoded

def decode(l):
    return ''.join([itos[i] for i in l])

# create the train and test splits
# Find the position of the last complete record in the first 90% of data
split_point = int(len(input_data) * 0.9)
# Look for the next <|end|> token after this point
end_token = "<|end|>\n"
next_end = input_data.find(end_token, split_point)
if next_end != -1:
    # Split after the end token (including the newline)
    split_point = next_end + len(end_token)
else:
    # Fallback if no end token is found
    print("Warning: Couldn't find clean split point. Using approximate split.")

train_data = input_data[:split_point]
val_data = input_data[split_point:]

# Verify the split is clean
print(f"Train data ends with: {train_data[-20:].replace('\n', '\\n')}")
print(f"Val data begins with: {val_data[:20].replace('\n', '\\n')}")

# encode both to integers
train_ids = encode(train_data)
val_ids = encode(val_data)
print(f"train has {len(train_ids):,} tokens")
print(f"val has {len(val_ids):,} tokens")

# export to bin files
train_ids = np.array(train_ids, dtype=np.uint16)
val_ids = np.array(val_ids, dtype=np.uint16)

meta = {
    'vocab_size': vocab_size,
    'itos': itos,
    'stoi': stoi,
}
print(meta)
```
